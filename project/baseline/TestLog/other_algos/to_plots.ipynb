```python
import pandas as pd 
import glob
final = pd.DataFrame()
for file in glob.glob("*crossover*"):
    final = pd.concat([final,pd.read_csv(file, header=None)], axis=0)
final.columns = ["0","1","2","3","4","5","6","7","Crossover operators","9","Mutation operators","11","12","13","14"]
#final["Algorithem"] = final["2"]  + final["6"].map(str) + final["7"] + final["8"] + final["9"].map(str) + final["10"] + final["11"].map(str) + final["11"].map(str) + final["12"].map(str) + final["13"].map(str)



import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize':(24,16)})
ax = sns.lineplot(x="3", y="5", data=final, hue="Crossover operators")
ax.set_title('Performance of Crossover operators', fontsize=25)
ax.legend(frameon=False, loc='lower center',ncol=2, fontsize=20)
ax.set_xlabel("number of generations",fontsize=20)
ax.set_ylabel("unseen accuracy",fontsize=20)
ax.tick_params(labelsize=18)
ax.legend(loc='best',fontsize='xx-large')
fig = ax.get_figure()
fig.savefig("Performance of Crossover operators.png") 


```
