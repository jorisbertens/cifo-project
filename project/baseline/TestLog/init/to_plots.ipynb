```python
import pandas as pd 
import glob
final = pd.DataFrame()
for file in glob.glob("*selection*"):
    final = pd.concat([final,pd.read_csv(file, header=None)], axis=0)
final.columns = ["0","1","2","3","4","5","6","Selection algorithems","8","9","10","11","12","13","14"]
#final["Algorithem"] = final["2"]  + final["6"].map(str) + final["7"] + final["8"] + final["9"].map(str) + final["10"] + final["11"].map(str) + final["11"].map(str) + final["12"].map(str) + final["13"].map(str)



import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize':(24,16)})
ax = sns.lineplot(x="3", y="5", data=final, hue="Selection algorithems")
ax.set_title('Performance of Selection algorithems', fontsize=25)
ax.legend(frameon=False, loc='lower center',ncol=2, fontsize=20)
ax.set_xlabel("number of generations",fontsize=20)
ax.set_ylabel("unseen accuracy",fontsize=20)
ax.tick_params(labelsize=18)
ax.legend(loc='best',fontsize='xx-large')
fig = ax.get_figure()
fig.savefig("Performance of Selection algorithems.png") 
```

```python
import glob
import itertools
final = pd.DataFrame()
for file in glob.glob("*variety.py_log1*"):
    final = pd.concat([final,pd.read_csv(file, header=None)], axis=0)
final.columns = ["0","1","Inititalization", "3","4","5","6","7","Crossover operators","9","Mutation operators","11","12","13","14","pheno entropy","geno entropy","pheno variance","geno variance"]
#final["Algorithem"] = final["2"]  + final["6"].map(str) + final["7"] + final["8"] + final["9"].map(str) + final["10"] + final["11"].map(str) + final["11"].map(str) + final["12"].map(str) + final["13"].map(str)

columns = ["pheno entropy", "pheno variance", "geno entropy", "geno variance"]
for column in columns:
    #final[column] = (final[column]-final[column].mean())/final[column].std()
    final[column]=(final[column]-final[column].min())/(final[column].max()-final[column].min())

import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize':(4,15)})
sns.set_style(style='white')

fig, axes = plt.subplots(nrows=4, sharex=True)
palette = itertools.cycle(sns.color_palette())


for index, column in enumerate(columns):
    sns.regplot(x="3", y=column, data=final,
                ax=axes[index], label=column,
                color=next(palette), scatter_kws={'s':10, 'alpha':0.6}, order=3)
    axes[index].set(xlabel='')
    
axes[index].set(xlabel='Number of generations')

fig.tight_layout()
    
#fig.savefig("Variety measures.png") 
fig.show()
```
